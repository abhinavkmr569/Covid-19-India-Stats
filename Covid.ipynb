### Importing Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

### Importing Dataset

```python
df = pd.read_json('https://www.mohfw.gov.in/data/datanew.json')
df
```

Removing the 1st and last row respectively

```python
df=df.iloc[:-1,1:]
```

```python
total_states = np.arange(len(df['state_name'])) # to be used for plotting    
total_states
```

### Stack All the columns in the bar chart

```python
df.columns
```

```python
df=df.set_index('state_name',drop=True)
```

```python
df.head()
```

```python
df.plot.barh(stacked=True,figsize=(16,10))
plt.yticks(total_states);
```

```python
a = df.index
```

```python
sns.set(style="whitegrid")
figure(figsize=(16,10),dpi = 160)
ax = sns.barplot(x=df['death'], y=a, data=df)

plt.xlabel('Number of Deaths')  
plt.ylabel('States')  
plt.title('Statewise Number of Deaths due to Covid-19 Cases');
```

```python
figure(figsize=(16,10),dpi = 160)
ax = sns.barplot(x=df['cured'], y=a, data=df)

plt.xlabel('Number of Patients Cured from Covid-19')  
plt.ylabel('States')  
plt.title('Statewise Number of Patients Cured from Covid-19');
```

```python
figure(figsize=(16,10),dpi = 160)
ax = sns.barplot(x=df['positive'], y=a, data=df)

plt.xlabel('Number of Positive Covid-19 Patients')  
plt.ylabel('States')  
plt.title('Statewise Number of Positive Covid-19 Patients');
```

```python
figure(figsize=(64,64),dpi = 320)
df.plot(kind='barh', stacked=True)
plt.yticks(total_states);
```
